### Importamos librerias

```python
import time
import matplotlib.pyplot as plt
import segunda_parte as sp
import numpy as np
```

### Definimos listas de archivos con casos de prueba

```python
archivos_catedra = [
    "./pruebas_catedra/5.txt",        
    "./pruebas_catedra/10.txt",  
    "./pruebas_catedra/20.txt", 
    "./pruebas_catedra/25.txt",  
    "./pruebas_catedra/50.txt",  
    "./pruebas_catedra/100.txt", 
    "./pruebas_catedra/1000.txt", 
    "./pruebas_catedra/2000.txt", 
    "./pruebas_catedra/5000.txt", 
    "./pruebas_catedra/10000.txt"
]

archivos_propios = [
    "./pruebas_propias/unico_elemento.txt",
    "./pruebas_propias/valores_crecientes.txt",
    "./pruebas_propias/valores_decrecientes.txt",
    "./pruebas_propias/aleatorios_1000.txt",
    "./pruebas_propias/aleatorios_10000.txt",
]
```

### Funciones auxiliares

```python
def parsearArchivo(archivo):
    with open(archivo, 'r') as archivo:
        for linea in archivo:
            if linea[0] != '#':
                arr = [int(num) for num in linea.split(';') if num.strip().isdigit()]
    return arr

def generar_casos(archivos):
    casos = []
    for archivo in archivos:
        casos.append(parsearArchivo(archivo))
    return casos

def medir_tiempos(casos):
    tiempos = []
    for monedas in casos:
        inicio = time.time()
        sp.juego_programacion_dinamica(monedas)
        fin = time.time()
        tiempos.append(fin - inicio)
    return tiempos

def calcular_tamanos(casos):
    tamanos = []
    for monedas in casos:
        tamanos.append(len(monedas))
    return tamanos
```

```python
casos_catedra = generar_casos(archivos_catedra)
tamanos_catedra = calcular_tamanos(casos_catedra)
tiempos_catedra = medir_tiempos(casos_catedra)

casos_propios = generar_casos(archivos_propios)
tamanos_propios = calcular_tamanos(casos_propios)
tiempos_propios = medir_tiempos(casos_propios)
```

## Graficamos

```python
plt.plot(tamanos_catedra, tiempos_catedra, label="Tiempo medido")
plt.xlabel("Tamano de la entrada")
plt.ylabel("Tiempo (s)")
plt.title("Medici√≥n de complejidad temporal")
plt.legend()
plt.show()
```

```python
coef = np.polyfit(tamanos_catedra, tiempos_catedra, 2)
ajuste = np.polyval(coef, tamanos_catedra)

# Graficar ajuste
plt.plot(tamanos_catedra, tiempos_catedra, 'o', label="Datos medidos")
plt.plot(tamanos_catedra, ajuste, '-', label="Ajuste cuadr√°tico")
plt.xlabel("Tamano de la entrada")
plt.ylabel("Tiempo (s)")
plt.title("Verificaci√≥n de complejidad O(n^2)")
plt.legend()
plt.show()
```

```python
plt.plot(tamanos_propios, tiempos_propios, label="Tiempo medido")
plt.xlabel("Tamano de la entrada")
plt.ylabel("Tiempo (s)")
plt.title("Medici√≥n de complejidad temporal")
plt.legend()
plt.show()
```

```python
coef = np.polyfit(tamanos_propios, tiempos_propios, 2)
ajuste = np.polyval(coef, tamanos_propios)

# Graficar ajuste
plt.plot(tamanos_propios, tiempos_propios, 'o', label="Datos medidos")
plt.plot(tamanos_propios, ajuste, '-', label="Ajuste cuadr√°tico")
plt.xlabel("Tamano de la entrada")
plt.ylabel("Tiempo (s)")
plt.title("Verificaci√≥n de complejidad O(n^2)")
plt.legend()
plt.show()
```
